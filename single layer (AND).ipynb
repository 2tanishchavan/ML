```python
import numpy as np

# Initialize weights and bias
weights = np.random.rand(2)
bias = np.random.rand()

# Define AND gate truth table
input_patterns = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
targets = np.array([0, 0, 0, 1])

# Activation function (threshold)
def activate(x):
    return 1 if x >= 0 else 0

# Single Layer Perceptron Learning algorithm
def perceptron_learning(input_pattern, target, weights, bias, learning_rate):
    output = activate(np.dot(input_pattern, weights) + bias)
    error = target - output
    weights += learning_rate * error * input_pattern
    bias += learning_rate * error
    return output, weights, bias

# Training loop
learning_rate = 0.1
epochs = 100
for epoch in range(epochs):
    for i in range(len(input_patterns)):
        output, weights, bias = perceptron_learning(input_patterns[i], targets[i], weights, bias, learning_rate)
        print(f"Input: {input_patterns[i]}, Target: {targets[i]}, Output: {output}, Weights: {weights}, Bias: {bias}")
```

```python
import numpy as np

# Define the inputs and outputs for the OR gate
inputs = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
outputs = np.array([0, 0, 0, 1])

# Initialize the weights randomly
weights = np.random.randn(2)

# Set the learning rate
learning_rate = 0.1

# Train the perceptron
for epoch in range(100):
    for i in range(len(inputs)):
        # Calculate the weighted sum of the inputs
        weighted_sum = np.dot(inputs[i], weights)

        # Apply the activation function (e.g., Heaviside step function)
        activation = 1 if weighted_sum > 1 else 0

        # Update the weights using the Perceptron learning rule
        error = outputs[i] - activation
        weights += learning_rate * error * inputs[i]

# Make predictions on new inputs
def predict(inputs):
    weighted_sum = np.dot(inputs, weights)
    activation = 1 if weighted_sum > 1 else 0
    return activation

print(predict([0, 0]))  # Output: 0
print(predict([0, 1]))  # Output: 0
print(predict([1, 0]))  # Output: 0
print(predict([1, 1]))  # Output: 1
```
