```python
# Step 1: Load the dataset
import pandas as pd

# Assuming 'Salary_Data.csv' contains the dataset
data = pd.read_csv('salary_data.csv')

# Step 2: Explore the data
print(data.head())
print(data.describe())
```

```python
# Step 3: No preprocessing needed assuming the dataset is clean

# Step 4: Split the data
from sklearn.model_selection import train_test_split

X = data.iloc[:, :-1].values  # Independent variable (years of experience)
y = data.iloc[:, 1].values    # Dependent variable (salary)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
# Step 5: Choose a model
from sklearn.linear_model import LinearRegression

# Step 6: Train the model
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
# Step 7: Evaluate the model
from sklearn.metrics import mean_squared_error, r2_score

y_pred = regressor.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("Mean Squared Error:", mse)
print("R-squared:", r2)
# Step 8: Make predictions
# Predict salary hike for a new employee with 5 years of experience
new_exp = [[5]]
predicted_salary = regressor.predict(new_exp)
print("Predicted Salary Hike for 5 Years of Experience:", predicted_salary)
```

```python
# Step 9: Visualize results (optional)
import matplotlib.pyplot as plt

plt.scatter(X_test, y_test, color='blue')
plt.plot(X_test, y_pred, color='red')
plt.title('Salary vs Experience (Test set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```
